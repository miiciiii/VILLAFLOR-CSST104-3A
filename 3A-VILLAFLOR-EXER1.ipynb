02 : Import the DFA Class from the automata.fa.dfa Module

```python
from automata.fa.nfa import NFA
```

03 : Define the NFA with its States


*   **States**
*   **Input Symbols**
*   **Transitions**
*   **Initials States**
*   **Final States**







```python
nfa = NFA(
    states = {'q0', 'q1', 'q2'},
    input_symbols = {'0', '1'},
    transitions = {
        'q0' : {'0' : {'q0', 'q1'}},
        'q1' : {'1' : {'q2'}},
        'q2' : {},
    },
    initial_state = 'q0',
    final_states = {'q0', 'q2'}
)
```

04 : Define a Functions is_accepted to Check if a Given Input String is Accepted by the DFA

```python
def is_accepted(input_str):
  return nfa.accepts_input(input_str)
```

05 : Test the DFA with a List of Strings

```python
test_strings = ['010', '1001', '001', '111', '0101']

for test_string in test_strings:
  if is_accepted(test_string):
    print(f'Strings "{test_string}" is accepted ')
  else:
    print(f'String "{test_string}" is not accepted ')
```
