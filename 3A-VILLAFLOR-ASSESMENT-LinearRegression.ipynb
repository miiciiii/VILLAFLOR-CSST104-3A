# 01 : Importing Libraries and Reading Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')

```

```python
df.columns
```

02 : Visualizing Plot

```python
x = df['Previous_Exam_Scores']
y = df['Grades']

plt.scatter(x, y)
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable (predictor)')
plt.title('linear')
plt.show()
```

# 03 : Getting Model Coefficents

```python
x = df[['x']]
y = df[['y']]


model = LinearRegression()

model.fit(x, y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print(f"""
slope : {slope}
intercept : {intercept}
""")
```

# 04 : Visualizing the Linear Regression Line

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color = 'green', marker = '+')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable (predictor)')
plt.title('Linear Regression')
plt.show()
```

# 05 : Making Predictions

```python
value_to_predict = [[np.random.randint(100, 200)]]

predicted_Y = model.predict(value_to_predict)

print(f"""
Value to Predict : {value_to_predict}
Predicted Value of Y : {predicted_Y}""")
```
