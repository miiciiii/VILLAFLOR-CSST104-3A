# 01 : Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

02 : Importing Dataset

```python
df = pd.read_csv(r'/content/Lesson_4_student_grades_dataset.csv')

print(type(df))
```

03 : Preparing Data for Modeling

```python
x = df[['Hours_of_Study', 'Attendance_Percentage', 'Previous_Exam_Scores',
       'Extracurricular_Activities', 'Travel_Time_to_School']]
y = df[['Grades']]
```

04 : Train and Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1, random_state = 50)
```

05 : Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

06 : Making Prediction

```python
y_predict = model.predict(x_test)
```

# 07 : Mean Squared Error

```python
mse = mean_squared_error(y_test, y_predict)
```

08 : Print Result

```python
new_data = np.array([[9.586459, 91.271388, 82.927026, 1	, 53.305077]])

prediction = model.predict(new_data)

grade = (prediction / 60) * 100



result = f"""
mean squared error : {mse}
prediction : {y_predict}
predicted grade : {prediction[0]}
grade : {grade}
"""

print(result)
```
