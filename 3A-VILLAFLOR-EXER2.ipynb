02 : import library

```python
from automata.fa.dfa import DFA
```

03 : Define DFA, States, Transitions, Start, Final, Input Symbols

```python
dfa = DFA(
    states = {'q0', 'q1'},
    input_symbols = {'0', '1'},
    transitions = {
        'q0' : {'0' : 'q0', '1' : 'q1'},
        'q1' : {'0' : 'q1', '1' : 'q0'}
    },
    initial_state = 'q0',
    final_states = {'q0'}
)
```

04 : Define Function is_accepted

```python
def is_accepted(input_str):
  return dfa.accepts_input(input_str)
```

05 : Test the DFA

```python
test_strings = ['1001', '1001001', '10101011', '1100010101', '1111']

for test_str in test_strings:
  if is_accepted(test_str):
    print(f'String "{test_str}" is accepted')
  else:
    print(f'String "{test_str}" is not accepted')
```
